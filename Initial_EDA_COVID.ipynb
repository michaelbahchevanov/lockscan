### 2.1 Loading in & glancing over the data

Firstly, we will import the libraries required for the task of data wrangling and visualization.

```python
df = pd.read_csv('./data/covid-data.csv')
```

Next we can look into duplicated or null values.

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

We can see that there are no duplicated values but there are some missing records from within the other columns. We will see which of them are relevant and we will find a way to fill them in.
***
Next we can look into the countries with the most and the countries with least amount of confirmed cases. We will limit that number to 10 for ease of visualization.

```python
grouping_cc = df.groupby(['location', 'continent', 'date'])['total_cases'].sum()
grouping_r = grouping_cc.groupby(['location', 'continent']).max()
df_f = pd.DataFrame(grouping_r.groupby(['location']).sum().sort_values(ascending=False))
df_f = df_f.fillna('NA')
highest_10 = df_f.head(10)
lowest_10 = df_f.tail(10)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12,8), sharey=True)
fig.suptitle('Confirmed COVID-19 Cases By Country', size=20)
sns.barplot(ax=ax[0],data=highest_10, x=highest_10.index, y='total_cases')
ax[0].ticklabel_format(style='plain', axis='y')
ax[0].set_title('Countries With Highest Confirmed Cases')
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation=60)

sns.barplot(ax=ax[1],data=lowest_10, x=lowest_10.index, y='total_cases')
ax[1].ticklabel_format(style='plain', axis='y')
ax[1].set_title('Countries With Lowest Confirmed Cases')
ax[1].set_ylabel('')
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=70)

fig.tight_layout()
```

In the graph above, we have visualized the countries with most and the countries with least amount of confirmed cases. By a huge margin, US has the most confirmed cases, albeit the pandemic began in China that is not in this graph.
***  
Now we can look into the `human_development_index` column which represent a composite index for the basic standard of living. We will be looking into the continents first.

```python
g = df.groupby(['location', 'continent'])['human_development_index'].sum().sort_values(ascending=False)
g1 = g.groupby(['continent']).mean().sort_values(ascending=False)
hdi_df = pd.DataFrame(g1)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(12, 8))
sns.barplot(x=hdi_df.index, y=hdi_df['human_development_index'])
ax.set_title('Continents by their Human Development Index')
ax.set_ylabel('HDI')
ax.set_xlabel('')
```

As observed, _Europe_ is the continent with the highest `human_development_index` and _Oceania_ is the lowest.  
***  
Now we can look into the countries with the highest and the lowest `human_development_index`.

```python
hdic_df = pd.DataFrame(g)
top_10 = hdic_df.head(10).reset_index().set_index('location')
least_10 = hdic_df.tail(10).reset_index().set_index('location')

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16,8), sharey=True)

fig.suptitle('Countries with Highest and Lowest HDI', fontsize=20)

sns.barplot(x=top_10.index, y=top_10['human_development_index'], ax=ax[0])
ax[0].set_xticklabels(ax[0].get_xticklabels(), rotation = 60)

sns.barplot(x=least_10.index, y=least_10['human_development_index'], ax=ax[1])
ax[1].set_xticklabels(ax[1].get_xticklabels(), rotation=60)
ax[1].set_ylabel(ylabel='')
```

```python
top_10['continent'].value_counts()
```

```python
least_10['continent'].value_counts()
```

The preceding chart shows the `human_development_index` in the countries with the highest and the lowest scores. We can also see the distribution between continents in the tables below.  
***
Next, we will be looking into the confirmed COVID-19 cases and the hospitalized patients daily. Data will be taken from the top 10 countries with highest amount of cases.

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(16,8))

fig.suptitle('Confirmed COVID-19 cases and hospitalizations per day', fontsize=16)

g = df.groupby(['date', 'location'])['total_cases'].sum().reset_index()
df_c = pd.DataFrame(g)
df_c = df_c[df_c['location'].isin(highest_10.index)]
df_c['date'] = pd.to_datetime(df_c['date'])

sns.lineplot(data=df_c, x=df_c['date'], y=df_c['total_cases'], hue='location', ax=ax[0])
ax[0].ticklabel_format(style='plain', axis='y')

g = df.groupby(['date', 'location'])['hosp_patients'].sum().reset_index()
df_c = pd.DataFrame(g)
df_c = df_c[df_c['location'].isin(highest_10.index)]
df_c['date'] = pd.to_datetime(df_c['date'])

sns.lineplot(data=df_c, x=df_c['date'], y=df_c['hosp_patients'], hue='location', ax=ax[1])
ax[1].ticklabel_format(style='plain', axis='y')
```

The plots above show us the top 10 countries with highest amount of cases and their respective values. We can observe the following:  
* The higher the amount of cases, the higher the hospitalized patient rates  
* There is missing data at places  
* At the beginning of the pandemic _(2020-03)_ there is a surge of hospitalizations

***  
We can further our observations by looking into the vaccination rates.

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(16,8))

fig.suptitle('People vaccinated (at least first dose) per day', fontsize=16)

g = df.groupby(['date', 'location'])['people_vaccinated'].sum().reset_index()
df_c = pd.DataFrame(g)
df_c = df_c[df_c['location'].isin(highest_10.index)]
df_c['date'] = pd.to_datetime(df_c['date'])

sns.lineplot(data=df_c, x=df_c['date'], y=df_c['people_vaccinated'], hue='location', ax=ax)
ax.ticklabel_format(style='plain', axis='y')
```

We can see that the __US__ has by far the fastest growth of vaccinated population (at least the first dose). But how does that match with the _confirmed cases_ daily.

```python
g = df.groupby(['date', 'location'])['people_vaccinated', 'total_cases'].sum().reset_index()
df_c = pd.DataFrame(g)
df_c = df_c[df_c['location'] == 'United States']
df_c['date'] = pd.to_datetime(df_c['date'])
df_cr = df_c[['date', 'people_vaccinated', 'total_cases']].resample('M', on='date').sum().reset_index(drop=False)
df_cr['date'] = pd.to_datetime(df_cr['date']).dt.date

plt.figure(figsize=(16,8))
plt.title('Vaccinated people and Confirmed cases for the US', fontsize=16)
ax = sns.barplot(data=df_cr, x=df_cr['date'], y=df_cr['total_cases'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=50)
ax.ticklabel_format(style='plain', axis='y')
ax1 = ax.twinx()
ax1.plot(ax.get_xticks(), df_cr['people_vaccinated'], marker='o')
ax1.ticklabel_format(style='plain', axis='y')
```

The plot above shows the _confirmed cases_ (bar), and _vaccinated people_, with the first dose at least (line), and we can see that the vaccination efforts prove fruitful. It is apparent that in _April_ the vaccination reaches a peak and the confirmed cases dip. We can also see a slowing down in cases since the beginning of vaccination in _January_.  
***  
Next, we will be looking into the stringency and how it changes based on different metrics. It would be helpful to "zoom-out" and look at the bigger picture, so we will be looking into the monthly data for the aggregation of the top 10 countries.

```python
g = df.groupby(['date', 'location'])['stringency_index', 'total_cases'].sum().reset_index()
df_c = pd.DataFrame(g)
df_c = df_c[df_c['location'].isin(highest_10.index)]
df_c['date'] = pd.to_datetime(df_c['date'])
df_cr = df_c[['date', 'stringency_index', 'total_cases']].resample('M', on='date').mean().reset_index(drop=False)
df_cr['date'] = pd.to_datetime(df_cr['date']).dt.date

plt.figure(figsize=(16,8))
plt.title('Stringency and Confirmed cases for the Top 10 Countries', fontsize=16)
ax = sns.barplot(data=df_cr, x=df_cr['date'], y=df_cr['stringency_index'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=50)
ax1 = ax.twinx()
ax1.plot(ax.get_xticks(), df_cr['total_cases'], marker='o')
ax1.ticklabel_format(style='plain', axis='y')
```

From the graph, we can see that the stringency (bar) and the confirmed cases (line) are following a pattern. When the stringency index is high _(above 70)_ the total cases are slowed down and when the index is lower, they go up.  
***  
Now we can look how the stringency relates to the hospitalizations.

```python
g = df.groupby(['date', 'location'])['stringency_index', 'hosp_patients'].sum().reset_index()
df_c = pd.DataFrame(g)
df_c = df_c[df_c['location'].isin(highest_10.index)]
df_c['date'] = pd.to_datetime(df_c['date'])
df_cr = df_c[['date', 'stringency_index', 'hosp_patients']].resample('M', on='date').mean().reset_index(drop=False)
df_cr['date'] = pd.to_datetime(df_cr['date']).dt.date

plt.figure(figsize=(16,8))
plt.title('Stringency and Hospitalization cases for the Top 10 Countries', fontsize=16)
ax = sns.barplot(data=df_cr, x=df_cr['date'], y=df_cr['stringency_index'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=50)
ax1 = ax.twinx()
ax1.plot(ax.get_xticks(), df_cr['hosp_patients'], marker='o')
ax1.ticklabel_format(style='plain', axis='y')
```

Looking at the stringency (bar) and hospitalizations (line), we can see a pattern. We observe the following:  
* In April, 2020, where the stringency peaked, there is a big drop in hospitalization cases in the following months  
* In June, 2020, where the stringency dropped, there is an increase in the hospitalization cases following

***  
With the graph above we should look into the vaccination efforts and the hospitalizations.

```python
g = df.groupby(['date', 'location'])['people_vaccinated', 'hosp_patients'].sum().reset_index()
df_c = pd.DataFrame(g)
df_c = df_c[df_c['location'].isin(highest_10.index)]
df_c['date'] = pd.to_datetime(df_c['date'])
df_cr = df_c[['date', 'people_vaccinated', 'hosp_patients']].resample('M', on='date').mean().reset_index(drop=False)
df_cr['date'] = pd.to_datetime(df_cr['date']).dt.date

plt.figure(figsize=(16,8))
plt.title('Vaccinated people and Hospitalization cases for the Top 10 Countries', fontsize=16)
ax = sns.barplot(data=df_cr, x=df_cr['date'], y=df_cr['hosp_patients'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=50)
ax1 = ax.twinx()
ax1.plot(ax.get_xticks(), df_cr['people_vaccinated'], marker='o')
ax1.ticklabel_format(style='plain', axis='y')
```

The preceding graph shows the vaccinated people, with at least first dose (line), and the hospitalized patients (bar). We can observe that the more people are vaccinated, the less hospitalizations there are, which shows that the vaccination efforts are succeeding in reducing the hospitalizations.
